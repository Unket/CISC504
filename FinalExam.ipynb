```python
#Develop and code a word parsing program which repalces certain grammatical elements in a phrase with user input as follows
# The octopus snagged the blue fish
our_str = 'The octopus snagged the blue fish'
subject = input("Enter replace for octopus ")
new_str = our_str.replace('octopus', subject)
adjective = input("Enter replace for blue ")
new_str = new_str.replace('blue', adjective)
objects = input("Enter replace for fish ")
new_str = new_str.replace('fish', objects)
verb = input("Enter replace for snagged ")
new_str = new_str.replace('snagged', verb)
print(new_str)
```

```python
#Develop and code a guessing program, which asks the user to generate a random number in the range (1:100). 
try :
    guessing = float(input('Enter number between 1 -100 : '))
    if guessing<1 or guessing>100 : print ('Number entered is out of range ')
    if guessing>=1 and  guessing<=50 : guess = guessing +1
    print ('New number is ',guess)
    if guessing>50 and  guessing<=100 : guess = guessing -1
    print ('New number is ',guess)
except ValueError:
    print ('Input parameter is not a number')
```
